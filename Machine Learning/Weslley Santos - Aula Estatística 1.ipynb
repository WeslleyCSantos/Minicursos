# Vari√¢ncia

* A vari√¢ncia mede o qu√£o "espalhado" o conjunto de dados est√°

* A vari√¢ncia $\sigma^2$ √© dada pela m√©dia das diferen√ßas quadradadas da m√©dia

* Complicado?

* Ex:
  
  Qual √© a vari√¢ncia do dataset (1, 4, 5, 4, 8)?

    * A m√©dia √© dada por: $(1+4+5+4+8)/5 = 4.4$
    * Agora achamos a diferen√ßa de cada valor para a m√©dia:  $(-3.4, -0.4, 0.6, -0.4, 3.6) $
    * Elevamos as diferen√ßas ao quadrado: $(11.56, 0.16, 0.36, 0.16, 12.96)$
    * Achamos a m√©dia das diferen√ßas ao quadrado: $\sigma ^ 2 =  (11.56 + 0.16 + 0.36 + 0.16 + 12.96) / 5 = 5.04$

# Desvio Padr√£o

* O desvio padr√£o √© a raiz quadrada da vari√¢ncia

  $\sigma ^2 = 5.04$

  $\sigma = \sqrt{5.04}= 2.24$

Logo, o desvio padr√£o de $(1+4+5+4+8)$ √© $2.24$

* Este m√©todo √© usado normalmente para identificar outliers. Valores de dados que est√£o a mais de um desvio padr√£o de dist√¢ncia da m√©dia podem ser considerados raros.

* Voc√™ pode falar o qu√£o extremo √© um ponto por quantos sigmas de dist√¢ncia ele est√° da m√©dia.

# Popula√ß√£o vs Amostra

* Se voc√™ est√° est√° trabalhando com uma amostra do conjunto de dados ao inv√©s do conjunto de dados inteiro ( toda a popula√ß√£o) 
  * Voc√™ dever√° utilizar a vari√¢ncia da amostra ao inv√©s da vari√¢ncia da popula√ß√£o
  * Para $N$ amostras, voc√™ divide as vari√¢ncias ao quadrado por $N-1$ ao inv√©s de $N$
  * No exemplo anterior, a vari√¢ncia da popula√ß√£o foi calculada como:

  $\sigma ^2 = (11.56 + 0.16 + 0.36 + 0.16 + 12.96) / 5 = 5.04 $

  * A vari√¢ncia da amostra seria dada por:

  $ùëÜ ^2= (11.56 + 0.16 + 0.36 + 0.16 + 12.96) / 4 = 6.3$

## Defini√ß√£o matem√°tica

* ## Vari√¢ncia da popula√ß√£o:
$\sigma ^2 = \frac{\sum {(X - \mu  )^2}}{N}$

* ## Vari√¢ncia da amostra:

## Distribui√ß√£o Uniforme

```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

## Distribui√ß√£o Normal/Gaussiana

```
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

Para gerar alguns pontos que seguem a distribui√ß√£o normal, mu √© a m√©dia desejada e sigma o desvio padr√£o

```
import numpy as np
import matplotlib.pyplot as plt

mu = 5.0
sigma = 2.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

## Fun√ß√£o de probabilidade exponencial

```
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

## Fun√ß√£o de massa binomial

```
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```
