# **Tipo de dado numérico**

* Representa uma medida quantitativa
   * Ex: Alturas, tempo de carregamento de uma página, preço de itens, etc.
* É divida em dois tipos:


* # Dados discretos
  * Baseados em números inteiros, servem para contar a quantidade de vezes que um evento ocorre
    * Quantas vezes eu ganhei na mega-sena esse ano?
    * Quantas vezes choveu quando eu pensei em ir pro IFPB?

* # Dados contínuos
  * Tem uma quantidade infinita de valores possíveis
    * Qual é a altura média da turma?
    * Qual é o meu CRE?



---


# **Tipo de dado categórico**

* Representa uma medida qualitativa, sem sentido matemático inerente
  * Ex: Gênero, raça, País de residência, Perguntas de Verdadeiro/Falso, etc.


* Podemos atribuir números para as categorias para representá-los de forma mais 
compacta, mas os números não terão sentido matemático.
---


# **Tipo de dado ordinal**

* É uma mistura de dados númericos e categóricos
* Trocando em miúdos, são dados categóricos que possuem sentido matemático
  * Ex: Notas de filmes em escala de 1-5
  * As notas devem ser 1, 2, 3, 4 ou 5
  * Mas os estes valores tem sentido matemático, uma nota 1 é pior que um 2.

# Média

* É o valor médio de um conjunto de dados
* Calculado por meio da soma dos valores dividido pelo número de amostras
* Ex:
  * Número de vezes que olhei no celular por hora:

    0, 2, 3, 2, 1, 0, 0, 2, 0

    ## **A média é dada por (0+2+3+2+1+0+0+2+0) / 9 = 1.11**

---

# Mediana

* É o valor do meio em um conjunto de dados ordenado

* Ex:

  # 0, 2, 3, 2, 1, 0, 0, 2, 0

  ## Ordenando

  # 0, 0, 0, 0, 1, 2, 2, 2, 3

 ##  O valor do meio é o 1, que é nosso valor mediano

* E se eu tiver um número par de amostras?

  A mediana será dada pela média dos dois valores centrais


* A mediana é menos sensível a outliers (Valores aberrantes) que a média

  * Ex: Nos EUA, a renda média anual é de 72.641 dólares, porém o valor mediano é de 51.939 dólares, mostrando que a média é enviesada pela quantidade de bilionários no país.

## Média vs mediana

Vamos criar dados falsos, centrados ao redor de 27000, com desvio padrão de 15000, com 10000 pontos de dados (Não se preocupe se não entender os termos, veremos mais a frente)

Depois vamos calcular a média, que deve estar próxima de 27000:

```
import numpy as np
np.random.seed(42)
salario = np.random.normal(27000,15000,10000)
np.mean(salario)
```

Podemos segmentar os dados de salário em 50 baldes e plotar em um histograma

```
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(salario, 50)
plt.show()
```

Agora vamos calcular a mediana, como a distribuição está bem equilibrada, o valor deve estar perto de 27000

```
np.median(salario)
```

Agora adicionamos um milionário aos salários.

```
salario = np.append(salario, [1000000000])
```

A mediana não mudará muito, mas a média sim

```
np.median(salario)
```

```
np.mean(salario)
```

# Variância

* A variância mede o quão "espalhado" o conjunto de dados está

* A variância $\sigma^2$ é dada pela média das diferenças quadradadas da média

* Complicado?

* Ex:
  
  Qual é a variância do dataset (1, 4, 5, 4, 8)?

    * A média é dada por: $(1+4+5+4+8)/5 = 4.4$
    * Agora achamos a diferença de cada valor para a média:  $(-3.4, -0.4, 0.6, -0.4, 3.6) $
    * Elevamos as diferenças ao quadrado: $(11.56, 0.16, 0.36, 0.16, 12.96)$
    * Achamos a média das diferenças ao quadrado: $\sigma ^ 2 =  (11.56 + 0.16 + 0.36 + 0.16 + 12.96) / 5 = 5.04$

# Desvio Padrão

* O desvio padrão é a raiz quadrada da variância

  $\sigma ^2 = 5.04$

  $\sigma = \sqrt{5.04}= 2.24$

Logo, o desvio padrão de $(1+4+5+4+8)$ é $2.24$

* Este método é usado normalmente para identificar outliers. Valores de dados que estão a mais de um desvio padrão de distância da média podem ser considerados raros.

* Você pode falar o quão extremo é um ponto por quantos sigmas de distância ele está da média.

# População vs Amostra

* Se você está está trabalhando com uma amostra do conjunto de dados ao invés do conjunto de dados inteiro ( toda a população) 
  * Você deverá utilizar a variância da amostra ao invés da variância da população
  * Para $N$ amostras, você divide as variâncias ao quadrado por $N-1$ ao invés de $N$
  * No exemplo anterior, a variância da população foi calculada como:

  $\sigma ^2 = (11.56 + 0.16 + 0.36 + 0.16 + 12.96) / 5 = 5.04 $

  * A variância da amostra seria dada por:

  $𝑆 ^2= (11.56 + 0.16 + 0.36 + 0.16 + 12.96) / 4 = 6.3$

## Definição matemática

* ## Variância da população:
$\sigma ^2 = \frac{\sum {(X - \mu  )^2}}{N}$

* ## Variância da amostra:
$ s ^2 = \frac{\sum {(X - M  )^2}}{N-1}$

# Exemplos de distribuição de dados

## Distribuição Uniforme

```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

## Distribuição Normal/Gaussiana

```
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

Para gerar alguns pontos que seguem a distribuição normal, mu é a média desejada e sigma o desvio padrão

```
import numpy as np
import matplotlib.pyplot as plt

mu = 5.0
sigma = 2.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

## Função de probabilidade exponencial

```
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

## Função de massa binomial

```
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```
